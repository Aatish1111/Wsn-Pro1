```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
num_nodes = 100  # Number of sensor nodes
grid_size = 100  # Grid size
initial_energy = 1.0  # Initial energy of each node
max_rounds = 100  # Number of rounds

# Initialize node properties
np.random.seed(0)
x_positions = np.random.uniform(0, grid_size, num_nodes)
y_positions = np.random.uniform(0, grid_size, num_nodes)
energies = np.full(num_nodes, initial_energy)

# Data structures to store results
rounds = []
num_messages_sent = []

# SPIN algorithm
for round in range(max_rounds):
    # Generate random data at each node
    data = np.random.randint(1, 100, num_nodes)
    
    # Calculate the number of messages sent in this round
    messages_sent = 0
    
    # Nodes negotiate and exchange data
    for sender_node in range(num_nodes):
        if energies[sender_node] <= 0:
            continue
        
        for receiver_node in range(num_nodes):
            if sender_node == receiver_node:
                continue
            
            if energies[receiver_node] <= 0:
                continue
            
            # Negotiation logic (for simplicity, just sending data if it's greater)
            if data[sender_node] > data[receiver_node]:
                data[receiver_node] = data[sender_node]
                messages_sent += 1
                energies[sender_node] -= 0.1  # Energy consumption for message transmission
    
    # Store results
    rounds.append(round)
    num_messages_sent.append(messages_sent)
    
    # Update energy levels for all nodes
    for node in range(num_nodes):
        energies[node] -= 0.01  # Energy consumption for idle listening

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(rounds, num_messages_sent, marker='o', linestyle='-')
plt.title('SPIN Algorithm Simulation')
plt.xlabel('Rounds')
plt.ylabel('Number of Messages Sent')
plt.grid(True)
plt.show()
```
